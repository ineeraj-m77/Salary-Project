```python
import pandas as pd
```

```python
df = pd.read_csv('glassdoor_jobs.csv')
```

```python
df.head(2)
```

```python
df.shape
```

### Sal Parsing, Company name text only
### State Field, age of company
### parsing of job description

```python
df = df[df['Salary Estimate'] != '-1']
```

```python
df.shape
```

```python
salary = df['Salary Estimate'].apply(lambda x : x.split('(')[0])
```

```python
salary
```

```python
remove_k_doll = salary.apply(lambda x : x.replace('K','').replace('$',''))
```

```python
remove_k_doll
```

```python
df['Hourly'] = df['Salary Estimate'].apply(lambda x : 1 if 'per hour' in x.lower() else 0)
```

```python
df['Hourly']
```

```python
df['Employer_Provided'] = df['Salary Estimate'].apply(lambda x : 1 if 'employer provided salary' in x.lower() else 0)
```

```python
df.head()
```

```python
minus_hr = remove_k_doll.apply(lambda x : x.lower().replace('per hour','').replace('employer provided salary:',''))
```

```python
minus_hr
```

```python
df['max-salary'] = minus_hr.apply(lambda x : int(x.split('-')[1]))
```

```python
df['min-salary'] = minus_hr.apply(lambda x : int(x.split('-')[0]))
```

```python
df.head()
```

```python
del df['mxn-salary']
```

```python
df['avg-salary'] = (df['max-salary'] + df['min-salary'])/2
```

```python
df['company_txt'] = df.apply(lambda x : x['Company Name'] if x['Rating'] <0 else x['Company Name'][:-3],axis = 1)
```

```python
df.shape
```

```python
df['State'] = df['Location'].apply(lambda x : x.split(',')[1])
```

```python
df.State.value_counts()
```

```python
df['Same_State'] = df.apply(lambda x : 1 if x.Location == x.Headquarters else 0, axis =1)
```

```python
df['age'] = df.Founded.apply(lambda x : x if x < 0 else 2020-x)
```

```python
df['python_yn'] = df['Job Description'].apply(lambda x : 1 if 'python' in x.lower() else 0)
```

```python
df['python_yn'].value_counts()
```

```python
df['r_yn'] = df['Job Description'].apply(lambda x : 1 if 'r studio' in x.lower() or 'r_studio' in x.lower() else 0)
```

```python
df['r_yn'].value_counts()
```

```python
df['spark_yn'] = df['Job Description'].apply(lambda x : 1 if 'spark' in x.lower() else 0)
df['spark_yn'].value_counts()
```

```python
df['aws_yn'] = df['Job Description'].apply(lambda x : 1 if 'aws' in x.lower() else 0)
df['excel_yn'] = df['Job Description'].apply(lambda x : 1 if 'excel' in x.lower() else 0)
```

```python
df['aws_yn'].value_counts()
```

```python
df['excel_yn'].value_counts()
```

```python
df.columns
```

```python
df_out = df.drop(columns = ['Unnamed: 0'])
```

```python
df_out.head()
```

```python
df_out.to_csv('salary_data_cleaned.csv' , index = False)
```
